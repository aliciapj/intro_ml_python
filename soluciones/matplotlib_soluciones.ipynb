```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.DataFrame({'x_data': range(1,21), 'y_data': np.random.randn(20), 'z_data': np.random.randint(100, size=20)})
```

```python
plt.plot('x_data', 'y_data', data=df)
plt.show()
```

```python
plt.bar('x_data', 'y_data', data=df)
plt.show()
```

```python
plt.scatter('x_data', 'y_data', data=df)
plt.show()
```

```python
plt.scatter('x_data', 'y_data', s='z_data', data=df)
plt.show()
```
